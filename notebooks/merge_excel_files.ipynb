```python
# Import required libraries
import sys
import os
from pathlib import Path
import pandas as pd

# Add the src directory to the Python path
sys.path.append('../src')

# Import our file merger
from file_merger import FileMerger

print("✅ Libraries imported successfully!")
print(f"📊 Pandas version: {pd.__version__}")
print(f"🐍 Python version: {sys.version}")
print(f"📁 Current working directory: {os.getcwd()}")
```

```python
# Set your data directory path here
# Example: data_directory = "/path/to/your/data"
# For this project, use the Data directory in the parent folder
data_directory = "../Data"

# Convert to Path object
data_path = Path(data_directory).resolve()

print(f"📂 Data directory: {data_path}")
print(f"📁 Directory exists: {data_path.exists()}")
print(f"📁 Is directory: {data_path.is_dir()}")

if data_path.exists() and data_path.is_dir():
    # List subdirectories
    subdirs = [d for d in data_path.iterdir() if d.is_dir()]
    print(f"📁 Found {len(subdirs)} subdirectories:")
    for subdir in subdirs:
        print(f"  - {subdir.name}")
else:
    print("❌ Directory not found or not accessible!")
```

```python
# Create file merger instance with column cleaning options
# You can customize these options:
column_cleaning_options = {
    'strip_whitespace': True,      # Remove leading/trailing spaces
    'normalize_spaces': True,     # Replace multiple spaces with single space
    'lowercase': False,           # Convert to lowercase
    'remove_special_chars': False, # Remove special characters
    'handle_common_variations': True # Handle common column name variations
}

merger = FileMerger(
    str(data_path),
    clean_columns=True,  # Set to False to disable column cleaning
    column_cleaning_options=column_cleaning_options
)

print("✅ File merger initialized!")
print(f"📂 Target directory: {merger.root_directory}")
print(f"🧹 Column cleaning: {'ENABLED' if merger.clean_columns else 'DISABLED'}")
print(f"📄 Supported formats: Excel (.xlsx, .xls) and CSV (.csv)")

if merger.clean_columns:
    print("🔧 Cleaning options:")
    for option, enabled in merger.column_cleaning_options.items():
        status = "✓" if enabled else "✗"
        print(f"  {status} {option}")

# Validate the directory
is_valid = merger.validate_directory()
print(f"✅ Directory validation: {'PASSED' if is_valid else 'FAILED'}")

if not is_valid:
    print("❌ Errors found:")
    for error in merger.errors:
        print(f"  - {error}")
```

```python
# Process all subdirectories
print("🚀 Starting Excel file merging process...")
print("=" * 60)

results = merger.process_all_subdirectories()

print("\n" + "=" * 60)
print("📊 PROCESSING COMPLETE!")
print("=" * 60)
```

```python
# Print detailed summary
merger.print_summary(results)

# Display results as a DataFrame for better visualization
import pandas as pd

summary_data = {
    'Metric': [
        'Success',
        'Subdirectories Processed',
        'Total Subdirectories',
        'Files Merged',
        'Skipped (Single File)',
        'Errors Encountered'
    ],
    'Value': [
        '✅ Yes' if results['success'] else '❌ No',
        results['processed_subdirs'],
        results.get('total_subdirs', 0),
        results['total_files_merged'],
        results.get('skipped_single_file', 0),
        len(results['errors'])
    ]
}

summary_df = pd.DataFrame(summary_data)
print("\n📋 SUMMARY TABLE:")
print(summary_df.to_string(index=False))

# Show errors if any
if results['errors']:
    print(f"\n⚠️  DETAILED ERRORS ({len(results['errors'])}):")
    for i, error in enumerate(results['errors'], 1):
        print(f"{i}. {error}")
else:
    print("\n🎉 No errors encountered!")
```
